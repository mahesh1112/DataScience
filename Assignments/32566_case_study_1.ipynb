```python
import numpy as np
import pandas as pd
from faker import Faker
```

```python
np.random.seed(42)
fake = Faker()
```

```python
no = 100
genders = np.random.choice(['Male',"Female"],size=no)
names = [fake.first_name_male() if gender == 'Male' else fake.first_name_female() for gender in genders]

student_data = {
    "Student_id": np.arange(1,no+1),
    "Name": names,
    "Gender": genders,
    "Age": np.random.randint(18,22,size = no)
}

marks = {
    "Student_id": np.arange(1,no+1),
    "DS": np.random.randint(50,100,size = no),
    "CS": np.random.randint(50,100,size = no),
    "ANN": np.random.randint(50,100,size = no),
    "NLP": np.random.randint(50,100,size = no)
}

df = pd.DataFrame(student_data).merge(pd.DataFrame(marks), on = "Student_id")
```

```python
df.head(10)
```

```python
subjects = list(marks.keys())[1:]
for subject in subjects:
    df.loc[df.sample(frac = 0.1).index,subject] = np.nan
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna({"DS":df["DS"].mean(),"CS":df["CS"].mean(),"ANN":df["ANN"].mean(),"NLP":df["NLP"].mean()},inplace = True)
```

```python
df.isnull().sum()
```

```python
df = df.astype({"DS":'int',"CS":"int","ANN":"int","NLP":"int"})
```
